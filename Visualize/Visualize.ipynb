```python
import matplotlib.pyplot as plt
import numpy as np 
import torch
import pickle
import pandas as pd
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
# a sample of imputed data 
imputation12000=np.load('impute12000/imputation0.npy')
original12000=np.load('impute12000/original0.npy')
mask12000 = np.load('impute12000/mask0.npy')

# a sample imputed by early iteration
imputation1200=np.load('impute1200/imputation0.npy')
original1200=np.load('impute1200/original0.npy')
mask1200 = np.load('impute1200/mask0.npy')

K = imputation12000.shape[-2] #feature
L = imputation12000.shape[-1] #time length
```

```python
dataind = 0 #change to visualize a different time-series sample

plt.rcParams["font.size"] = 16
plt.subplots(nrows=9, ncols=4,figsize=(24.0,12.0))

for k in range(K):
    df = pd.DataFrame({"x":np.arange(0,L), "val":original12000[dataind,k,:],"y":mask12000[dataind,k,:]})
    df2 = pd.DataFrame({"x":np.arange(0,L), "val":imputation12000[dataind,k,:], "y":mask12000[dataind,k,:]})
    df2 = df2[df2.y == 0]

    plt.subplot(4,4,k+1)
    plt.plot(df.x,df.val, color = 'b',marker = 'o', linestyle='None')
    plt.plot(df2.x,df2.val, color = 'r',marker = 'x', linestyle='None')
```
